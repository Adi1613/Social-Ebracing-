## k-Means Clustering

```python
# kmeans :

from sklearn.cluster import KMeans

wcss = []
centroids = []


for i in range(1, 15):
    
    kmeans = KMeans(n_clusters=i, init='k-means++')
    kmeans.fit(normalized_df.values)
    wcss.append(kmeans.inertia_)
    centroids.append(i)
```

### Plotting the elbow plot

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(15, 5))
plt.plot(centroids, wcss)
plt.scatter(centroids, wcss, s=70, c='red')
plt.xlabel('no of clusters')
plt.ylabel('weighted clusters sum of squares')
plt.title('wcss vs no_of_clusters')
#plt.legend(label = "WCSS value")
```

### Number of optimal clusters are 5

```python
kmean_optimal = KMeans(n_clusters=5)
kmean_optimal.fit(normalized_df.values)
normalized_df['pred_kmeans_cluster'] = kmean_optimal.predict(normalized_df.values)
```

### Davies Bouldin Score

```python
from sklearn.metrics import davies_bouldin_score
davies_bouldin_score(normalized_df.iloc[:, :-1], normalized_df['pred_kmeans_cluster'])
```

## Hierarchical Clustering

```python
X = normalized_df.iloc[:,[0,1,2,3,4]].values
```

```python
# Using the dendrogram to find the optimal number of clusters
# Create the figure and set it's size.
fig = plt.figure(figsize=(16,10))
plt.rcParams.update({'font.size': 22})
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
```

### Number of optimal clusters shown by dendrogram is 2

```python
# Fitting Hierarchical Clustering to the dataset
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

```python
normalized_df['pred_agg_cluster'] = y_hc
```

### Visualizing the algomerative clustering results

```python
fig = plt.figure(figsize=(16,10))
sns.pairplot(data=normalized_df, hue='pred_agg_cluster')
```
