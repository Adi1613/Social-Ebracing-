```python
from mpl_toolkits.basemap import Basemap
```

```python
import pandas as pd
df0 = pd.read_csv('clust00.csv')
df0.head(5)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches

fig = plt.figure(figsize = (9,6))

map = Basemap(projection='cyl',
           llcrnrlat = -90,
           urcrnrlat = 90,
           llcrnrlon = -180,
           urcrnrlon = 180,
           resolution = 'c')
    
#map.drawparallels(np.arange(-90,90,10),labels=[True,False,False,False])
#map.drawmeridians(np.arange(-180,180,30),labels=[0,0,0,1])
    
map.drawcoastlines()
map.fillcontinents(color = '#0000ff', lake_color = '#FFFFFF')
map.drawcountries(color = 'black')
map.drawmapboundary(fill_color = '#FFFFFF')#'#00FFCD')


#plt.show()
#plt.savefig('test.png')

sites_lat_y0 = df0['Lat'].tolist()
sites_lon_x0 = df0['Long'].tolist()

x, y = map(sites_lon_x0, sites_lat_y0)

colors = ['red']#, '', 'yellow', 'red', 'blue', 'orange']
#colors = df0['pca_kmeans']
for i in range(len(sites_lat_y0)):
    clust0 = plt.scatter(x,y, s=50,marker='o',c = colors, zorder=5)#, c=colors)#, , alpha=1, edgecolor='k', linewidth=1, zorder=2)

red_patch = mpatches.Patch(color='red', label='Cluster 0')
plt.legend(handles=[red_patch], loc = 4)
#plt.legend(handles=[clust0])
plt.title('Cluster 0', fontsize=20)
    
#plt.show()

#plt.savefig('Cluster00.png')
```
